```python
#Starting with model creation
from sklearn.model_selection import train_test_split
```

```python
#Converting the string date to date-time in the 'indian_cases' dataset
import datetime as dt
indian_cases['date']=pd.to_datetime(indian_cases['date'])
indian_cases.head()
```

```python
#Converting the string date to ordinal in the 'indian_cases' dataset
indian_cases['date']=indian_cases['date'].map(dt.datetime.toordinal)
indian_cases.head()
```

```python
#Assigning the dependant and independant variable for the model
x=indian_cases['date']
y=indian_cases['total_cases']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#Starting with linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#Helping the model to learn from the training data
import numpy as np
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
x_test
```

```python
#Predicting for the testing values
y_pred=lr.predict(np.array(x_test).reshape(-1,1))
```

```python
#Printing average squared difference between the estimated values and the actual value
from sklearn.metrics import mean_squared_error
mean_squared_error(x_test,y_pred)
```
